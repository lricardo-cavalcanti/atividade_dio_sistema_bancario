```python
menu = """

[d] Depositar
[s] Sacar
[e] Extrato
[nu] Novo usuário
[nc] Nova conta
[q] Sair

=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3
usuarios = []
contas = []
numero_conta_existente = 1

def depositar(valor, saldo, extrato):
    if valor > 0:
        saldo += valor
        extrato += f"Depósito: R$ {valor:.2f}\n"
        print(f"Depósito de R${valor:.2f} realizado com sucesso!")

    else:
        print("Operação falhou! O valor informado é inválido.")

    return saldo, extrato

def sacar(saldo, valor, extrato, limite, numero_saques, LIMITE_SAQUES):

    excedeu_saldo = valor > saldo
    excedeu_limite = valor > limite
    excedeu_saques = numero_saques >= LIMITE_SAQUES

    if excedeu_saldo:
        print("Operação falhou! Você não tem saldo suficiente.")

    elif excedeu_limite:
        print("Operação falhou! O valor do saque excede o limite.")

    elif excedeu_saques:
        print("Operação falhou! Número máximo de saques excedido.")

    elif valor > 0:
        saldo -= valor
        extrato += f"Saque: R$ {valor:.2f}\n"
        numero_saques += 1
        print(f"Saque de R${valor:.2f} realizado com sucesso!")
    else:
        print("Operação falhou! O valor informado é inválido.")

    return saldo, extrato, numero_saques

def mostrar_extrato(saldo, /, extrato):
    print("\n================ EXTRATO ================")
    print("Não foram realizadas movimentações." if not extrato else extrato)
    print(f"\nSaldo: R$ {saldo:.2f}")
    print("==========================================")

def novo_usuario():
    nome = input("Digite o nome completo: ")
    data_nascimento = input("Digite a data de nascimento (dd/mm/aaaa): ")
    cpf = input("Digite o CPF (somente números): ")
    cpf = ''.join(filter(str.isdigit, cpf))

    if len(cpf) != 11:
      print("CPF inválido. Digite apenas os números do CPF.")
      return

    if any(usuario['cpf'] == cpf for usuario in usuarios):
        print("Já existe um usuário cadastrado com esse CPF.")
        return

    logradouro = input("Digite o logradouro: ")
    bairro = input("Digite o bairro: ")
    cidade = input("Digite a cidade: ")
    sigla_estado = input("Digite a sigla do estado: ")

    endereco = {
        "logradouro": logradouro,
        "bairro": bairro,
        "cidade": cidade,
        "sigla_estado": sigla_estado
    }

    usuario = {
        "nome": nome,
        "data_nascimento": data_nascimento,
        "cpf": cpf,
        "endereco": endereco
    }

    usuarios.append(usuario)
    print("Usuário cadastrado com sucesso!")

def nova_conta():
    cpf = input("Digite o CPF do usuário: ")
    cpf = ''.join(filter(str.isdigit, cpf))

    if len(cpf) != 11:
        print("CPF inválido. Digite apenas os números do CPF.")
        return

    usuario = next((u for u in usuarios if u['cpf'] == cpf), None)
    if not usuario:
        print("Usuário não encontrado.")
        return

    global numero_conta_existente
    conta = {
        "agencia": "0001",
        "numero_conta": numero_conta_existente,
        "usuario": usuario
    }
    contas.append(conta)
    numero_conta_existente += 1
    print("Conta criada com sucesso!")


while True:

    opcao = input(menu)

    if opcao == "d": #deposito
        valor = float(input("Informe o valor do depósito: "))

        saldo, extrato = depositar(valor, saldo, extrato)

    elif opcao == "s": #sacar
        valor = float(input("Informe o valor do saque: "))

        saldo, extrato, numero_saques = sacar(saldo, valor, extrato, limite, numero_saques, LIMITE_SAQUES)

    elif opcao == "e": #extrato
        mostrar_extrato(saldo, extrato)

    elif opcao == "nu":
        novo_usuario()

    elif opcao == "nc":
        nova_conta()

    elif opcao == "q":
        break

    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
